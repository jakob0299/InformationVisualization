```python
# ---------- Diagnostics + robust concat ----------
import numpy as np

keys = ["Country Name", "Code", "Year"]   # canonical keys we use


def ensure_keys_and_types(df):
    # unify column names & strip whitespace
    df = df.rename(columns=lambda c: c.strip())
    # unify 'year' vs 'Year'
    if 'year' in df.columns and 'Year' not in df.columns:
        df = df.rename(columns={'year':'Year'})
    # drop Unnamed cols (from CSV)
    unnamed = [c for c in df.columns if c.startswith("Unnamed")]
    if unnamed:
        df = df.drop(columns=unnamed)
    # ensure keys exist and are strings
    for k in keys:
        if k in df.columns:
            df[k] = df[k].astype(str)
    return df

cleaned = [ensure_keys_and_types(df.copy()) for df in dfs]

# 1) Print duplicates diagnostics
print("Duplicates diagnostics (count of duplicate key-rows per DF):")
for i, df in enumerate(cleaned):
    dup_count = df.duplicated(subset=keys).sum() if all(k in df.columns for k in keys) else "missing keys"
    rows = len(df)
    print(f" DF[{i}] rows={rows:6d}, duplicate_key_rows={dup_count}")

# 2) Resolve duplicates where present
resolved = []
for i, df in enumerate(cleaned):
    if not all(k in df.columns for k in keys):
        raise ValueError(f"DF[{i}] missing one of keys {keys} -> columns: {df.columns.tolist()}")
    if df.duplicated(subset=keys).any():
        # choose aggregation: numeric -> mean, others -> first non-null
        agg_funcs = {}
        for col in df.columns:
            if col in keys:
                continue
            if pd.api.types.is_numeric_dtype(df[col]):
                agg_funcs[col] = 'mean'
            else:
                # custom: take first non-null; fallback to first if all null
                agg_funcs[col] = lambda s: s.dropna().iloc[0] if s.dropna().size else s.iloc[0]
        # groupby to collapse duplicates
        df = df.groupby(keys, dropna=False).agg(agg_funcs).reset_index()
        print(f" DF[{i}] duplicates collapsed -> new rows: {len(df)}")
    resolved.append(df)

# 3) now set index and concat
dfs_indexed = [df.set_index(keys) for df in resolved]

# final concat (sort=False to avoid ordering warnings)
merged = pd.concat(dfs_indexed, axis=1, join="outer", sort=False).reset_index()

# 4) remove exact duplicate column names (keeps first)
merged = merged.loc[:, ~merged.columns.duplicated()]

# 5) convert Year back to integer where possible
merged['Year'] = pd.to_numeric(merged['Year'], errors='coerce').astype('Int64')

print("\nFinal merged shape:", merged.shape)
print("Columns (sample):", merged.columns.tolist()[:30])

pd.set_option("display.max_rows", 10000)

merged = merged.sort_values(["Country Name", "Year"])
merged = merged.set_index(["Country Name", "Year"])

merged.head(1000)
# merged
```
